```python
# Import the required libraries and dependencies
import pandas as pd
import datetime as dt
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import hvplot.pandas
%matplotlib inline
```

```python
#reading in our data
data_df = pd.read_csv('healthcare_dataset.csv')
```

```python
#Cleaning the data and comparing them to values above to see if a lot of data was erased. As well as converting datatypes to datetime if needed.
data_df = data_df.dropna()
data_df['Date of Admission'] = pd.to_datetime(data_df['Date of Admission'])
data_df['Discharge Date'] = pd.to_datetime(data_df['Discharge Date'])
print(data_df)
```

```python
#What are the admission trends over time? We could slice the data even further if we see a specific timeframe that sticks out, in 2020 
#there was most likely a spike from COVID admissions etc.. so it could be needed to extrapolate further.

heatmap_df = data_df[['Date of Admission','Medical Condition']].set_index('Date of Admission')
heatmap_df.head()
cleaned_df = heatmap_df.groupby('Date of Admission')['Medical Condition'].size().reset_index()

cleaned_df.set_index('Date of Admission', inplace=True)

cleaned_df.hvplot.heatmap(x='index.year', y='index.month', C='Medical Condition', cmap='blues')
```

```python
#Expirementing 
admissions_by_months = data_df.resample('M', on='Date of Admission').size().reset_index()
heatmap_info = admissions_by_months.rename(columns={0: 'Number of Admissions'})
heatmap_info.head()
```

```python
#What are the admission trends over time? We could slice the data even further if we see a specific timeframe that sticks out, in 2020 
#there was most likely a spike from COVID admissions etc.. so it could be needed to extrapolate further.
#This sample here is a line plot instead of a heat map above for another visualization to help more if needed...

#Resample data to convert daily data into yearly data to help create this visualization.
admissions_trend = data_df.resample('Y', on='Date of Admission').size()
admissions_trend.plot(title='Number of Admissions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Admissions')
plt.show()
```

```python
# Question: What is the distribution of medical conditions among patients? With this data we could come to conclusions that a certain condition is
#spiking compared to our older data. And possibly create a time series analysis for medical condition's diagnosis' over time.
condition_counts = data_df['Medical Condition'].value_counts()
condition_counts.plot(kind='bar', title='Distribution of Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Patients')
plt.show()
```

```python
#Grouping the data by medical condition and calculate the average billing amount.
#This code below first groupby medical condition as the index and then compares to the billing amount so we have a filtered dataframe to plot/visualize
med_cond_billing = data_df.groupby('Medical Condition')['Billing Amount'].mean().sort_values()

print(med_cond_billing)
# Plotting the data
plt.figure(figsize=(10, 6))
med_cond_billing.plot(kind='bar', color='skyblue')
plt.title('Average Billing Amount by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#A boxplot chart can help visualize quartiles, quantiles, etc.
plt.figure(figsize=(12, 6))
data_df.boxplot(column='Billing Amount', by='Medical Condition', grid=False, vert=False)
plt.title('Billing Amount Distribution by Medical Condition')
plt.xlabel('Billing Amount')
plt.ylabel('Medical Condition')
plt.suptitle('')
plt.show()
```

```python
#This violiln plot can help visualize data points
plt.figure(figsize=(12, 6))
sb.violinplot(x='Billing Amount', y='Medical Condition', data=data_df, palette='muted')
plt.title('Billing Amount Distribution by Medical Condition')
plt.show()
```

```python
#What is the average billing amount compared to gender and medical condition?
#this data could be useful to insurance companies and patients medical history so they know what type of insurance a patient needs and their appropriate
#rate and risk to the insurance company
avg_billing = data_df.groupby(['Gender', 'Medical Condition'])['Billing Amount'].mean().unstack()
avg_billing.plot(kind='bar', title='Average Billing Amount by Gender and Medical Condition')
plt.xlabel('Gender')
plt.ylabel('Average Billing Amount')
plt.show()
```

```python
#Average age among patients compared to medical conditions? what conclusion can you make from this? you could see early warning signs of rising disease/
#rates among certain age groups and perhaps put out health warnings, information, etc. For instance, If we see obesity rising in younger adults we
#could employ physical education awarness, inform policy makers about our data, etc.
plt.figure(figsize=(12, 6))
data_df.boxplot(column='Age', by='Medical Condition', grid=False, vert=False)
plt.title('Age Distribution by Medical Condition')
plt.xlabel('Age')
plt.ylabel('Medical Condition')
plt.suptitle('')
plt.show()
```

```python
#This visualization was made in part from chatgpt and sb documentation to find different visualizations.
# Create age bins
data_df['Age Group'] = pd.cut(data_df['Age'], bins=range(0, 101, 10))

# Create a pivot table to aggregate data by Age Group and Medical Condition
age_condition_counts = data_df.pivot_table(index='Age Group', columns='Medical Condition', aggfunc='size', fill_value=0)
print(age_condition_counts)
# Plotting the grouped bar plot
fig, ax = plt.subplots(figsize=(14, 8))

# Number of bars per group
n_conditions = len(age_condition_counts.columns)


# Array of bar positions for each condition
positions = np.arange(len(age_condition_counts))

# Plot each medical condition as a separate set of bars
for i, condition in enumerate(age_condition_counts.columns):
    ax.bar(positions + i * width, age_condition_counts[condition], width=width, label=condition)

# Setting labels and title
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of Patients')
ax.set_title('Number of Patients by Age Group and Medical Condition')
ax.set_xticks(positions + width * (n_conditions / 2 - 0.5))
ax.set_xticklabels(age_condition_counts.index, rotation=45)
ax.legend(title='Medical Condition', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
