```python
# Import the required libraries and dependencies
import pandas as pd
import datetime as dt
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#reading in our data
data_df = pd.read_csv('healthcare_dataset.csv')
```

```python
data_df.value_counts
```

```python
#printing data
print(data_df)
print(data_df.describe())
```

```python
print(data_df.shape)
```

```python
#checking for null values in a heatmap
plt.figure(figsize = (8,6))
sb.heatmap(data_df.isnull(), cbar=False , cmap = 'magma')
```

```python
#Cleaning the data and comparing them to values above to see if a lot of data was erased.
data_df = data_df.dropna()
data_df['Date of Admission'] = pd.to_datetime(data_df['Date of Admission'])
data_df['Discharge Date'] = pd.to_datetime(data_df['Discharge Date'])
print(data_df)
```

```python
data_df.isnull().sum()
```

```python
data_df.value_counts()
```

```python
#grouping the data by medical condition and calculate the average billing amount
med_cond_billing = data_df.groupby('Medical Condition')['Billing Amount'].mean().sort_values()

print(med_cond_billing)
# Plotting the data
plt.figure(figsize=(10, 6))
med_cond_billing.plot(kind='bar', color='skyblue')
plt.title('Average Billing Amount by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question: What is the distribution of medical conditions among patients?
condition_counts = data_df['Medical Condition'].value_counts()
condition_counts.plot(kind='bar', title='Distribution of Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# What are the admission trends over time?

admissions_trend = data_df.resample('M', on='Date of Admission').size()
admissions_trend.plot(title='Monthly Admissions Trend')
plt.xlabel('Date')
plt.ylabel('Number of Admissions')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data_df.boxplot(column='Billing Amount', by='Medical Condition', grid=False, vert=False)
plt.title('Billing Amount Distribution by Medical Condition')
plt.xlabel('Billing Amount')
plt.ylabel('Medical Condition')
plt.suptitle('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sb.stripplot(x='Billing Amount', y='Medical Condition', data=data_df, jitter=True, palette='muted', alpha=0.7)
plt.title('Individual Billing Amounts by Medical Condition')
plt.show()
```

```python
g = sb.FacetGrid(data_df, col="Medical Condition", col_wrap=3, height=4, aspect=1.5)
g.map(plt.hist, "Billing Amount", bins=20, color='skyblue')
g.set_titles("{col_name}")
g.fig.suptitle('Billing Amount Distribution by Medical Condition', y=1.05)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sb.violinplot(x='Billing Amount', y='Medical Condition', data=data_df, palette='muted')
plt.title('Billing Amount Distribution by Medical Condition')
plt.show()
```

```python
#this data could be useful to insurance companies and patients medical history so they know what type of insurance a patient needs and their appropriate
#rate and risk to the insurance companypo
avg_billing = data_df.groupby(['Gender', 'Medical Condition'])['Billing Amount'].mean().unstack()
avg_billing.plot(kind='bar', title='Average Billing Amount by Gender and Medical Condition', legend=)
plt.xlabel('Gender')
plt.ylabel('Average Billing Amount')
plt.show()
```

```python
#age among patients compared to medical conditions? what conclusion can you make from this? you could see early warning signs of rising disease/
#rates among certain age groups and perhaps put out health warnings, information, etc.
plt.figure(figsize=(12, 6))
data_df.boxplot(column='Age', by='Medical Condition', grid=False, vert=False)
plt.title('Age Distribution by Medical Condition')
plt.xlabel('Age')
plt.ylabel('Medical Condition')
plt.suptitle('')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sb.violinplot(x='Medical Condition', y='Age', data=data_df)
plt.title('Age Distribution by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sb.swarmplot(x='Medical Condition', y='Age', data=data_df)
plt.title('Age Distribution by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
